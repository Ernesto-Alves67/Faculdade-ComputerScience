## Detecção de Rostos Humanos em Imagens

```python
import cv2

# Carregar o classificador pré-treinado para detecção de faces
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')

# Carregar a imagem do Google Drive
image_path = 'images/pessoas.jfif'
image = cv2.imread(image_path)

# Verificar se a imagem foi carregada corretamente
if image is None:
    print('Não foi possível carregar a imagem')
    exit()

# Converter a imagem para escala de cinza
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detectar faces na imagem
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1, minSize=(30, 30))

# Desenhar retângulos ao redor das faces detectadas
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Exibir a imagem com as faces detectadas
cv2.imshow('Face Recognition', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

# Histograma de Imagem

```python

import matplotlib.pyplot as plt

# Carregar a imagem
image = cv2.imread('images/pessoas.jfif')

# Converter a imagem para escala de cinza
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calcular o histograma
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

# Plotar o histograma
plt.figure()
plt.title('Histograma de Escala de Cinza')
plt.xlabel('Intensidade de Pixel')
plt.ylabel('Número de Pixels')
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```
